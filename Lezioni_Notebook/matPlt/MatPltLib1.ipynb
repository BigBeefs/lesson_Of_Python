```python
import matplotlib.pyplot as plt
import numpy as np
from numpy.random import rand 
```

```python
x = rand(100)     #Crea in maniera random dei numeri da 0,0 a 1
x
```

```python
y = rand(100)
y
```

```python
x = rand(100)
y = rand(100)               #Creo uno plot a puntini con la coppia coordinate di x e y
plt.scatter(x,y)
#plt.show    #Viene utilizzato nei file .py per poter visualizzarlo
```

```python
x = rand(1000)
y = rand(1000)               
plt.scatter(x,y,color='green')
```

```python
plt.plot([1, 2, 3, 4],'-r')  #Quando non esprimo uno delle due coordinate, viene improvvisato dal sistema nella maniera più appropriata
plt.ylabel('label y')        #diamo il nome all'asse  
```

```python
plt.plot([1, 2, 3, 4], 'or')  #Con "or" o "-r" selezioniamo la rappresentazione dei punti
plt.ylabel('label y')
```

mathplotlib è una libreria  modulare, posso aggiungere divesi moduli come plot label, possiamo dare le dimensioni o la rappresentazione grafica

```python
plt.plot([5,2,9,4],linewidth=3, marker='o', color = 'green')
plt.xlabel('Asse X')
plt.ylabel('Asse Y')
plt.title('Grafico')
```

```python
plt.plot([1,2,3,4],[1,4,9,16],'og') #Il dato 'og' indica lo stile della rappresentazione grafica dei punti
plt.plot([1,2,5,6],[4,5,9,7],'--r')
```

```python
x = np.linspace(0,10,100) #Il terzo data indica il numero di punti da rappresentare, in questo caso da 0 a 10 ci saranno 100 punti
y = np.exp(-x) #np.exp = funzione esponenziale
plt.plot(x,y) #Viene generato y in automatico e intelligente
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z= np.cos(x)
plt.plot(x,y)

```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z= np.cos(x)

plt.figure(figsize=(18,10))  #Crea una figura e la ridimensiona

plt.plot(x,y,'--')
plt.plot(x,z,'-')
plt.savefig('test.png') #Salviamo la figura 
```
