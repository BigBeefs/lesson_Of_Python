```python
import numpy as np
import streamlit as st
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv('Company.csv')
df
```

```python
df.drop(columns='Sales')
```

```python
X=df.drop(columns='Sales')
y=df['Sales']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25,random_state=667)
```

```python
X_train.shape,X_test.shape
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
res_df = pd.DataFrame(data=list(zip(y_pred, y_test)),columns=['predicted', 'real'])  #Creo un dataframe con i contenuti delle previsioni 
res_df
```

```python
res_df['error'] = res_df['real'] - res_df['predicted']
res_df 
```

```python
res_df.round(1)
```

```python
res_df['error'].mean()
```
